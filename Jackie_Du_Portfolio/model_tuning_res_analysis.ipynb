```python
# load packages 
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt

import time
# import xgboost
```

```python
# read in pickles 
pds = []

for i in range(36):
    filename = 'data/' + str(i) + '_subset_cv_res_v2.pickle'
    df_i = pd.read_pickle(filename)
    pds.append(df_i)
```

```python
df = pd.concat(pds)
```

```python
df.groupby('model').agg({'mean_cv_F1_score':'mean', 'std_cv_F1_score':'mean','val_f1':'mean'}).reset_index()
```

```python
df.groupby(['resampling_method', 'neighbors']).agg({'mean_cv_F1_score':'mean', 
                                     'std_cv_F1_score':'mean', 
                                     'val_f1':'mean'})
```

```python
df.groupby('subset').agg({'mean_cv_F1_score':'mean', 
                          'std_cv_F1_score':'mean',
                          'val_f1':'mean'}).sort_values('mean_cv_F1_score', ascending=False)
```
